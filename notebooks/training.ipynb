```python
# imp features
cols = [
    'person_income',
    'loan_amnt',
    'loan_percent_income',
    'loan_int_rate',
    'loan_grade',
    'cb_person_default_on_file',
    'cb_person_cred_hist_length',
    'person_emp_length',
    'person_home_ownership',
    'loan_intent',
    'person_age',
    'loan_status'
]
```

```python
df = df[cols]
```

# Handle missing values

```python
df["loan_int_rate"].fillna(df["loan_int_rate"].median(), inplace=True)
df["person_emp_length"].fillna(df["person_emp_length"].median(), inplace=True)
```

```python
df.duplicated().sum() # 165
df.drop_duplicates(inplace=True)
```

# Adding features

```python
df["is_high_debt"] = (df["loan_percent_income"] > 0.4).astype(int)
```

# Outlier detection

```python
df = df[(df["person_age"] < 60) & (df["person_emp_length"] < 30)]
```

```python
df.head()
```

# Preprocessing

```python
from sklearn.preprocessing import OrdinalEncoder, OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
```

```python
bin_col = ["cb_person_default_on_file"]
oe_col = ["loan_grade"]
ohe_col = ["person_home_ownership", "loan_intent"]
num_col = [
    'person_income',
    'loan_amnt',
    'loan_percent_income',
    'loan_int_rate',
    'cb_person_cred_hist_length',
    'person_emp_length',
    'person_age',
    'is_high_debt'
]
```

```python
preprocessing = ColumnTransformer([
    ("ohe", OneHotEncoder(drop="first", handle_unknown="ignore"), ohe_col),
    ("oe", OrdinalEncoder(categories=[["A","B","C","D","E","F","G"]]), oe_col),
    ("bin", OrdinalEncoder(), bin_col),
    ("num", StandardScaler(), num_col)
])
```

```python
X = df.drop(columns="loan_status")
y = df["loan_status"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
```

```python
lr_model = Pipeline([
    ('prep', preprocessing),
    ('clf', LogisticRegression(max_iter=1000))
])

lr_model.fit(X_train, y_train)
```

```python
rf_model = Pipeline([
    ("prep", preprocessing),
    ("clf", RandomForestClassifier(
    n_estimators=300,
    max_depth=12,
    min_samples_split=10,
    min_samples_leaf=5,
    max_features='sqrt',
    random_state=42,
    n_jobs=-1
    ))
])

rf_model.fit(X_train, y_train)
```

```python
from xgboost import XGBClassifier
```

```python
xgb_model = Pipeline([
    ("prep", preprocessing),
    ("clf", XGBClassifier(
        n_estimators=300,
        max_depth=6,
        learning_rate=0.05,
        subsample=0.8,
        colsample_bytree=0.8,
        eval_metric='logloss',
        random_state=42
    ))
])

xgb_model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, roc_auc_score

print("lr acc = ",accuracy_score(y_test, lr_model.predict(X_test)))
print("lr roc-auc = ",roc_auc_score(y_test, lr_model.predict_proba(X_test)[:, 1]))

print("rf acc = ",accuracy_score(y_test, rf_model.predict(X_test)))
print("rf roc-auc = ",roc_auc_score(y_test, rf_model.predict_proba(X_test)[:, 1]))

print("xgb acc = ",accuracy_score(y_test, xgb_model.predict(X_test)))
print("xgb roc-auc = ",roc_auc_score(y_test, xgb_model.predict_proba(X_test)[:, 1]))
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'clf__n_estimators': [200, 300, 500],
    'clf__max_depth': [4, 6, 8],
    'clf__learning_rate': [0.03, 0.05, 0.1],
    'clf__subsample': [0.8, 1.0],
    'clf__colsample_bytree': [0.8, 1.0]
}

grid = GridSearchCV(
    estimator=xgb_model,
    param_grid=param_grid,
    cv=5,
    scoring='roc_auc',
    n_jobs=-1,
    verbose=2
)

grid.fit(X_train, y_train)
```

```python
print("Best Params:", grid.best_params_)
print("Best AUC:", grid.best_score_)
```

```python
best_model = grid.best_estimator_

y_prob = best_model.predict_proba(X_test)[:,1]
print("Test AUC:", roc_auc_score(y_test, y_prob))
```

```python
best_model.named_steps['clf'].feature_importances_
```

```python
from sklearn.metrics import classification_report

for t in [0.3, 0.4, 0.5, 0.6]:
    y_pred = (y_prob > t).astype(int)
    print(f"\nThreshold = {t}")
    print(classification_report(y_test, y_pred))
```

```python
THRESHOLD = 0.4

def predict_loan(model, df):
    prob = model.predict_proba(df)[:,1]
    pred = (prob > THRESHOLD).astype(int)
    return pred, prob
```

```python
income = 50000
loan = 18000
lpi = loan / income

new_customer_df = pd.DataFrame([{
    "person_income": income,
    "loan_amnt": loan,
    "loan_percent_income": lpi,
    "loan_int_rate": 14,

    "loan_grade": "D",
    "cb_person_default_on_file": "N",

    "cb_person_cred_hist_length": 4,
    "person_emp_length": 2,
    "person_home_ownership": "RENT",
    "loan_intent": "PERSONAL",

    "person_age": 28,                
    "is_high_debt": int(lpi > 0.4)
}])
```

```python
pred, prob = predict_loan(best_model, new_customer_df)

print("Probability:", prob[0])
print("Decision:", "Reject" if pred[0]==1 else "Approve")
```
